```python
import tensorflow as tf
import pandas as pd
import numpy as np
import gc
import math
from gensim import corpora, models, utils
from gensim.models import LdaModel, CoherenceModel
import jieba

df = pd.read_csv("C:/Users/veniv/OneDrive/桌面/PLOITICAL INFORMATION/10802_2020party.csv")
df["Post_Date"] = df["Date"].str.split(" ").str.get(0).str.title() 
df_party = df[['Message','Post_Date']] 
```

```python
df_party = df_party.dropna()
df_party['Message']
```

# 斷詞

```python
doc_names = range(len(df_party))
text_list = list(df_party['Message'])
df_party_docs = pd.DataFrame(columns=['jieba_results'])
df_party_docs['jieba_results'] = df_green_docs['jieba_results'].astype('str')
```

```python
#停用字和自訂字典
jieba.load_userdict("C:/Users/veniv/cc/KCC Data/Dict/KccDict2020.txt")
stopwords = [line.strip() for line in open('C:/Users/veniv/cc/KCC Data/Dict/KccStopWord2020.txt', 'r', encoding='utf-8').readlines()]  

for i in doc_names:
    
    words = jieba.cut(text_list[i], cut_all = False)
    text = ''
    for word in words:
        if word not in stopwords: 
            text = text + ' ' + str(word)
    s = pd.Series({'jieba_results': text})
    df_party_docs = df_party_docs.append(s, ignore_index=True)  
    
df_party.reset_index(inplace=True)  
df_party['jieba_results'] = df_party_docs
df_party['df_class'] = 'green'
```

```python
party_file = df_party['jieba_results']
party_file
```

```python
party_list = []
for file in party_file:
            party_list.append(list(utils.simple_preprocess(file, deacc=True, min_len=2,)))
```

```python
dict = corpora.Dictionary(party_list)
print("dict = ", dict)
print("------------------------------")
```

```python
corpus = [dict.doc2bow(text) for text in party_list]   

print("corpus[:1] = ", corpus[:1])   
print("------------------------------")

print("print lists of [word:frequency]")
[[print((dict[id], freq)) for id,freq in cp] for cp in corpus[:1]]
print("================end of the ouput================")
```

# LDA

```python
lda_model = models.ldamodel.LdaModel(corpus=corpus, id2word= dict, num_topics=10, passes=20, alpha='auto', chunksize=50, per_word_topics=True)
output = open("C:/Users/veniv/cc/KCC Data/LDA_HW", 'w', encoding='utf-8-sig')


lstTM = lda_model.print_topics(num_topics=10, num_words=20)
for tupleTM in lstTM:
        str1 = "".join(str(x) for x in tupleTM)
        #print("type(str1) = ", type(str1))
        str1 += "\n"
        print("strTM = ", str1)
        output.write(str1)
        
output.close()
 
print("------------------------------")
print("Perplexity = ", lda_model.log_perplexity(corpus))

print("------------------------------")
cm = CoherenceModel(model=lda_model, texts=list(party_file), dictionary=dict, coherence= 'c_v' )
coher_lda = cm.get_coherence()
print("\n Coherence Score: ", coher_lda) 
```
